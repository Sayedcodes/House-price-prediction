### Cart System



```python
# cart
cart={
    "product_name":[],
    "product_quantity_list":[]
}
# add to cart function
def add_to_cart(product_name, product_quantity):
    #add product name
    cart["product_name"].append(product_name)
    #add product quantity
    cart["product_quantity_list"].append(product_quantity)
    print("Product added to cart successfully!")
```

```python
add_to_cart("Pc",1)
```

```python
# remove from cart

def remove_item(product_name):
    if product_name in cart["product_name"]:
        index= cart["product_name"].index(product_name)
        cart["product_name"].pop(index)
        print(f"{product_name} has been removed from the cart")
    else:
        print(f"{product_name} is not found in the cart")

```

```python
remove_item("Pc")
```

```python
# Update cart
def update_cart(product_name , product_quantity):
    if product_name in cart["product_name"]:
        index= cart["product_name"].index(product_name)
        cart["product_quantity_list"][index]= product_quantity
        print(f"{product_name},{product_quantity} has been updated in the cart")
    else:
        print(f"{product_name} is not found in the cart")
```

```python
update_cart("pc",3)
```

```python
# view cart
def view_cart():
    print("\n--- Cart Items ---")
    for i in range(len(cart["product_name"])):
        print(f"{cart['product_name'][i]}: {cart['product_quantity_list'][i]}")
```

```python
view_cart()
```

```python
# cart menu
def cart_menu():
    while True:
        print("\n--- Cart Menu ---")
        print("1: Add to Cart")
        print("2: Remove Item")
        print("3: Update Cart")
        print("4: View Cart")
        print("5: Exit")

        choose= input("Choose an option (1-5): ")
        if choose == "1":
            product_name = input("Enter product name: ")
            product_quantity = int(input("Enter product quantity: "))
            add_to_cart(product_name, product_quantity)
        elif choose == "2":
            product_name = input("Enter product name to remove: ")
            remove_item(product_name)
        elif choose == "3":
            product_name = input("Enter product name to update: ")
            product_quantity = int(input("Enter new product quantity: "))
            update_cart(product_name, product_quantity)
        elif choose == "4":
            view_cart()
        elif choose == "5":
            print("Exiting cart menu.")
            print("\n Thank you for using the cart system!")
            print("\n Goodbye!")
            break
        else:
            print("Invalid option. Please try again.")
# Run cart menu
cart_menu()

```

## Question
##### write a fuction to calculate frequency of char in string
##### programme --> p:1 , r:2 , g:3 , m:

```python
def frequency_counter(word):
    data={}
    for char in word:
        if char not in data.keys():
            data[char] = 1
        else:
            data[char] += 1
    return data
```

```python
x= frequency_counter("Sayed Mohammad Hamza")
for i,j in x.items():
    print(f"{i}: {j}")
```

### Kwargs Function

### Return function Usecase:

In Python, a return statement is used inside a function to send a value back to the place where the function was called.

Once return is executed, the function stops running, and any code after it is ignored.

```python
def add(a, b):
    return a + b  # Sends the sum back to the caller

result = add(5, 3)  # result gets the value 8
print(result)       # Output: 8
```

* return the value and store in variable for future use
* use with functions
* return a value from function for future
