## File Handling

 File handling in Python refers to the process of interacting with files on a computer's file
 system, allowing for operations such as creating, opening, reading, writing, appending,
 renaming, and deleting files.

## Key Concepts and Operations:

### Write a File

* open --> open the file then now you can read , write .etc....
* then write the file name what did you want to use
* then give the mode what type of mode you want to use like....
* --> w for write / r for read / a for append / b for binary and many more etc
* then in the next line use the variable which is used for create the file with the (.)methods like "write" 
* then close the file as you see in the code


```python
file= open("file_handling_text_check.txt","w") #create a file if not exists
file.write("This is my first file handling program in python") #write text to file
file.close() #close the file
```

### Read the File

* Read a file write the same name
* same way to read the file as like write the file but now use the (.read)method
* then print that variable
* last close the file

##### (.read) is method which can read all the text from file as you see in the code
* use when You want the full file at once
* output come in str(string)

```python
file= open("file_handling_text_check.txt","r") #open the file in read mode
data= file.read() #read the file
print(data) #print the data
# file.close() #close the file
```

##### (.readline) is method which can read a single line text from the file as you see in the code
* use when You want to read a sing line from data
* output come in list

```python
file= open("file_handling_text_check.txt","r") #open the file in read mode
data= file.readline() #read a single line from the file
print(data) #print the data
# file.close() #close the file
```

##### (.readlines) is method which can read a text line by line from the file as you see in the code
* use when You want to work line by line
* output come in list

```python
file= open("file_handling_text_check.txt","r") #open the file in read mode
data= file.readlines() #read all lines from the file
print(data) #print the data
# file.close() #close the file
```

### Question
Create a file and store the data and also read

```python
#data
cart= {"phone":30,"pc":40}

#create a file
cart_file= open("cart_data.txt","w+")
cart_file.write(str(cart)) #write the cart data to the file
cart_file.seek(0) #move the cursor to the beginning of the file
#why we use the seek? because after writing the cursor is at the end of the file 
# so we need to move it to the beginning to read the data
print(cart_file.read()) #read the cart data #print the cart data
cart_file.close() #close the file
```

management,

```python
with open("python_logo.jpg","rb") as img: #open the image file in read binary mode
   img.read() #read the image file
```
