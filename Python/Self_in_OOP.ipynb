Addmission form with the help of Constructor

```python
class Student_admission:
    institue_name="ITV"
    batch="d313"

    def __init__(self,name,phone_no,course):
        self.name= name
        self.ph= phone_no
        self.course= course
        self.fee= 0
    
    def show_details(self):
        print(f'''Student Details: \n
              Student Name: {self.name} \n
              Phone Number: {self.ph} \n
              Course Enrolled: {self.course} \n
              Course Fee: {self.fee}''')
        
    def fee_update(self,fee):
        self.fee= fee
        print("Addmission Done")
```
